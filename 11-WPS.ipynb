# WPS

Los pasos para la configuración del código fuente de WPS son:

1. Descargar y desempaquetar el código fuente
1. Definir variables para el uso de netCDF, HDF5 y Jasper
1. Definir la variable WRF
1. Editar el script __configure__ 
1. Ejecutar el script __configure__  y seleccionar el tipo de compilación dmpar
1. Editar el archivo de configuración __configure.wps__ para el uso de los compiladores de intel MPI
1. Ejecutar el script __compile__  y guardar un registro  

<h2>Estos pasos deben ejecutarse en la terminal</h2>

## 1. Descargar y desempaquetar código fuente

~~~bash
WPS=https://github.com/wrf-model/WPS/archive/refs/tags/v4.3.1.tar.gz
cd $HOME/tmp/2022/wrf
curl -L $WPS | tar xzvf -
cd $HOME/tmp/2022/wrf/WPS-4.3.1
ls -F
~~~

## 2. Definir variables para el uso de compiladores, netCDF, HDF5 y Jasper
~~~bash
SETVARS_ARGS="--config=/opt/software/apps/intel/oneapi/config.txt"

source /opt/software/apps/intel/oneapi/setvars.sh 

export NETCDF=$HOME/tmp/2022/wrf

export HDF5=$HOME/tmp/2022/wrf

export JASPERINC=$HOME/tmp/2022/wrf/include

export JASPERLIB=$HOME/tmp/2022/wrf/lib

export PATH=$HOME/tmp/2022/wrf/bin:$PATH
~~~ 

## 3. Definir la variable WRF_DIR
Durante la configuración de WPS, requiere la ubicación del directorio de compilación de WRF en la variable WRF_DIR
~~~bash
export WRF_DIR=$HOME/tmp/2022/wrf/WRF-4.3.3
~~~

## 4. Editar el script configure

> Error reportado antes de editar el script __configure_

> Your Fortran + NETCDF did not run successfully.                                                                                                                               

~~~bash
vim +390 configure 
~~~

Contenido original de la línea 388 del archivo __configure__
~~~bash
./fort_netcdf > /dev/null 2>&1
~~~

Cambiar a

~~~bash
env LD_LIBRARY_PATH=$LD_LIBRARY_PATH:${NETCDF}/lib ./fort_netcdf > /dev/null 2>&1
~~~

## 6. Editar el archivo de configuración configure.wps para el uso de los compiladores de intel MPI

* línea 65  cambiar valor de la variable __DM_FC__ a __mpiifort__
* línea 66  cambiar valor de la variable __DM_CC__ a __mpiicc__

> antes

~~~bash
DM_FC               = mpif90
DM_CC               = mpicc
~~~

> después 

~~~bash
DM_FC               = mpiifort
DM_CC               = mpiicc
~~~
